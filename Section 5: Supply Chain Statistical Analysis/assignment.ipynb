```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Pineapple_juice.csv')
df.head()
```

```python
df.shape
```

```python
import scipy.stats as st
from scipy.stats import normaltest, kstest
```

```python
demand = np.array(df['Pineapple juice'])
```

```python
st.normaltest(demand)
```

```python
st.kstest(demand,'norm',args=(demand.mean(),demand.std()))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = np.array(df['Price']).reshape(-1,1)
y = np.array(df['Pineapple juice'])
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
print(model.coef_, model.intercept_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X,y)
```

```python
df[['Price','Pineapple juice']].corr()
```

```python
fig, ax1 = plt.subplots(figsize=(100, 10))
ax1.plot(df['Day'], df['Price'], label='Line 1', color='blue', linestyle='-', marker='o')
ax2 = ax1.twinx()
ax2.plot(df['Day'], df['Pineapple juice'], label='Line 2', color='red', linestyle='--', marker='s')
```

```python
new_price = np.random.randint(10,50,10).reshape(-1,1)
```

```python
time_period = np.arange(10)
new_demand = model.predict(new_price)
fig, ax1 = plt.subplots(figsize=(8, 5))
ax1.plot(time_period, new_price, label='Line 1', color='blue', linestyle='-', marker='o')
ax2 = ax1.twinx()
ax2.plot(time_period, new_demand, label='Line 2', color='red', linestyle='--', marker='s')
```

## 2. List & list indexing

```python
# Create a list of prime numbers
prime_numbers = [1, 3, 5, 7, 11, 13, 17]
print(prime_numbers)
```

```python
# Indexing with list of numbers

# Index the second element of the list because the index starts from 0
print(prime_numbers[1])

# Index the first element of the list
print(prime_numbers[0])

# Index from the first to the third element of the list
print(prime_numbers[0:3])

# Index from the first to the fourth element of the list
print(prime_numbers[0:4])

# Index ignore the last element of the list
print(prime_numbers[0:-1])

# Index all element of the list
print(prime_numbers[:])

# Index from the fourth element to the end of the list
print(prime_numbers[3:])
```

```python
# Indexing with list of strings

# Create a list of strings
class_list = ['Mark', 'Jack', 'Mo', 'Raj', 'Sarah', 'Leo']
print(class_list)

# Index from the fourth element to the end of the list
print(class_list[3:])

# Index from the first to the third element of the list
print(class_list[:3])

# Index from the first to the third element of the list
print(class_list[0:3])
```

## 6. Boolean

```python
# Assign value for each variable
haytham = 32
mariam = 23
frank = 15
leo = 12
sarah = 8
```

```python
# Check true or false
print(haytham > 32)
print(haytham < 32)
print(haytham >= 32)
print(haytham <= 32)
print((haytham > 23) and (haytham < 40))
print((haytham > 23) & (haytham < 40))
print((haytham > 23) or (mariam > 26))
print((haytham > 23) | (mariam > 26))
print((haytham > 23) and (mariam == 23) and (leo < 15))
print(
    (haytham > 23) & 
      (mariam == 23) & 
      (leo < 15)
    )
```

## 7. Functions & branching

```python
# Create a function to check the status of a person
def status(person):
    if person >= 20:
        return 'adult'
    elif person >= 14:
        return 'teenager'
    else:
        return 'child'
```

```python
# Run the function
print(status(haytham))
print(status(mariam))
print(status(leo))
```

## 8. Apply the function using map()

```python
# Create a list
class_list = [haytham, mariam, leo, frank, sarah]
```

```python
# Using map the iterate the function to each element in the list
map_object = map(status, class_list)
status_list = list(map_object)
status_list
```

## 9. "for" loop

```python
# For each element (called "age") in the class_list, print the element
for age in class_list:
    print(age)

# For each element (called "i") in the class_list, print the element

for i in class_list:
    print(i)
```

```python
# Apply the function status() to each element in the list
for age in class_list:
    print(status(age))
```

```python
# Apply the function status() to each element in the list and store the result in a new list
status_list = []

for age in class_list:
    status_list.append(status(age))

status_list
```

## 10. Apply the function using map() on a dataframe

```python
# Define a function to check if the country is UK or not
def uk_or_not(country):
    if country == 'United Kingdom':
        return True
    else:
        return False
```

```python
# Apply the function uk_or_not()
print(uk_or_not('France'))
print(uk_or_not('United Kingdom'))
```

```python
# Using map() to apply the function to the dataframe
retail_data['uk'] = retail_data['Country'].map(uk_or_not)

# Check the result
retail_data['uk'].value_counts()
```

## 11. Using "for" loop on a dataframe

```python
# Index the first 10 rows of the data
first_10 = retail_data.iloc[:10,:]
first_10
```

```python
# Create a new column called 'uk1' and assign NaN to all rows
first_10['uk1'] = np.nan
first_10
```

```python
# Using for loop to apply the function uk_or_not() to each row in the column
for i in range(len(first_10)):
    first_10.loc[i, 'uk1'] = uk_or_not(first_10.Country[i])

# Check the result
first_10.head()
```

## Which airline is the most punctual?

```python
fl.columns
```

```python
punctual_carrier = fl[['carrier','dep_delay','arr_delay']].fillna(value = 0)
punctual_carrier
```

```python
punctual_carrier['not_punctual_time'] = punctual_carrier['dep_delay'] + punctual_carrier['arr_delay']
```

```python
punctual_carrier = punctual_carrier.groupby(by = 'carrier').agg(mean_delay = ('not_punctual_time', 'mean')).sort_values('mean_delay', ascending=True).reset_index()
punctual_carrier
```

```python
pd.merge(punctual_carrier, al, how = 'inner',on = 'carrier')[['carrier','name','mean_delay']]
```

## What destination is having the longest flight duration?

```python
fl['air_time'].describe()
```

```python
fl_at = fl.groupby(['origin','dest']).agg(average_airtime = ('air_time','mean')).reset_index().sort_values(by = 'average_airtime', ascending = False)
fl_at
```

```python
pd.merge(fl_at, ap[['faa','name']], left_on = 'dest', right_on = 'faa')
```

## Which airline is the worst in terms of delays

```python
pd.merge(punctual_carrier, al, how = 'inner',on = 'carrier')[['carrier','name','mean_delay']].sort_values(by = 'mean_delay', ascending = False)
```

## Which airline has the highest capacity of seats

```python
carrier = fl[['carrier','tailnum']].drop_duplicates()
carrier
```

```python
carrier_pl = pd.merge(carrier, pl[['tailnum','seats']], how = 'left', on = 'tailnum').fillna(0).groupby(by = 'carrier').agg(total_seats = ('seats','sum')).sort_values(by = 'total_seats', ascending = False)
carrier_pl
```

```python
carrier_pl = pd.merge(carrier_pl, al, on = 'carrier')
carrier_pl
```

## Which airplane model is the highest in use and from which manufacturer

```python
pl_use = fl.groupby('tailnum').agg(count_use = ('tailnum', 'count')).reset_index()
pl_use
```

```python
pd.merge(pl_use, pl[['tailnum', 'model', 'manufacturer']]).groupby(['model', 'manufacturer']).agg(total_flight = ('count_use', 'sum')).sort_values('total_flight', ascending = False).reset_index()
```

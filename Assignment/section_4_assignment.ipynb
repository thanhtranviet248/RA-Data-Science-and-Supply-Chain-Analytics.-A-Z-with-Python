```python
# Part 1

# Import package
import pandas as pd
import numpy as np

# Import data
df = pd.read_csv('cars.csv')

# Check number of row
print(f'''Number of row: {df.shape[0]}''')

# Check number of column
print(f'''Number of column: {df.shape[1]}''')

# Check unique value of column
print(f'''Unique value of the cyclenders: {df['cylenders'].unique()}''')
print(f'''Number of unique value of the cyclenders: {len(df['cylenders'].unique())}''')

# Mean of horsepower
print(f'''Mean of horsepower: {df['horsepower'].mean()}''')

# Max of horsepower
print(f'''Max of horsepower: {df['horsepower'].max()}''')

# Print the row with column "name" and "Price" where the price is the highest
print(f'''Car name with highest price: {df[df.Price == df.Price.max()][['name','Price']]}''')

# Rename column name
df.rename(columns={'name':'car_name'}, inplace = True)

# Check the first 5 row
df.head()
```

```python
# Part 2

# Index all row with column "car_name" and "Price"
car_pricing = df.loc[:,['car_name','Price']]
print(f'''
      Dataframe after indexing
      {car_pricing.head()}''')

# Define a function to categorize the car based on price
def pricing_category(price):
    if price < 20000:
        return 'Budget Car'
    elif price >= 20000 and price <= 35000:
        return 'Suitable Car'
    else:
        return 'Expensive Car'
    
# Create a new column to categorize the car based on price
car_pricing['category'] = car_pricing['Price'].apply(pricing_category)
print(f'''
      Dataframe with cateogry column
      {car_pricing.head()}''')

# Count the number of car in each category
print(f'''
      Number of car in each category {car_pricing['category'].value_counts()}''')
```

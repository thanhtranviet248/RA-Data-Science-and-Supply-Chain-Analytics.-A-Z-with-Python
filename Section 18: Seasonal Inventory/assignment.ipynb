```python
import pandas as pd
import inventorize3 as inv
```

```python
df = pd.read_csv('twentyeleven.csv')
df.head()
```

```python
df = df.groupby(by = ['Description']).agg(total_demand = ('Quantity', 'sum'), price = ('Price', 'mean')).reset_index()
df['cost'] = 0.4*df['price']
df['sd_demand'] = 0.1*df['total_demand']
```

```python
result = pd.DataFrame()

for i in range(len(df)):
    a = inv.MPN_singleperiod(mean = df['total_demand'][i], 
                             standerddeviation = df['sd_demand'][i], 
                             p = df['price'][i], c = df.cost[i], b = 0, g = 0)
    b = pd.DataFrame(a, index = [0])
    b['Description'] = df['Description'][i]
    result = pd.concat([result, b], axis = 0)
```
